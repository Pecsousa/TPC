### TPC6: Análise de dados: doença cardíaca

Descarregue o ficheiro de dados: [diabetes_prediction_dataset.csv](./datasets/diabetes_prediction_dataset.csv.zip).

Encontra mais informação sobre este conjunto de dados [aqui](https://www.kaggle.com/datasets/iammustafatz/diabetes-prediction-dataset).

Crie um programa em Python, conjunto de funções, que responda às seguintes questões:
* Crie uma função que lê a informação do ficheiro para um modelo, previamente pensado em memória;
* Crie uma função que calcula a distribuição da doença por sexo (tenha atenção que indivíduos doentes e não doentes no dataset);
* Crie uma função que calcula a distribuição da doença por escalões etários. Considere os seguintes escalões: [0-10], [11-24], [25-29], [30-34], [35-39], [40-44], ...
* Crie uma função que calcula a distribuição da doença por níveis de glucose. Considere um nível igual a um intervalo de 10 unidades, comece no limite inferior e crie os níveis necessários até abranger o limite superior;
* Crie uma função que imprime na forma de uma tabela uma distribuição;
* Especifique um programa que ao executar apresenta as tabelas correspondentes às distribuições pedidas.

```python
informacao=[]
file=open("diabetes_prediction_dataset.csv")
diab=0     

for linha in file:
    reader=linha.split(",")
    if "1" in reader[-1]:
        genero, idade, hipertensao, coracao = reader[0], float(reader[1]), reader[2], reader[3]
        fumador, bmi, HbA1c, glucose, diabetes=reader[4], float(reader[5]), float(reader[6]), int(reader[7]), reader[8]
        informacao.append((genero,idade,hipertensao,coracao,fumador,bmi,HbA1c,glucose,diabetes))
        diab+=1

def DistribuicaoGenero(): 
        F=0
        M=0
        for termo in informacao:
                if termo[0]=="Female":
                    F = F + 1
                elif termo[0] =="Male":
                    M = M + 1
        pM=round((M/diab)*100, 2)
        pF=round((F/diab)*100, 2)

        print('                       Masculino:' + str(pM) + '%''                              Feminino:' + str(pF) + '%')
    
def DistribuicaoIdade():
    intervalos=[(0,10),(11,24),(25,29),(30,34),(35,39),(40,44),(45,49),(50,54),(55,59),(60,64),(65,69),(70,74),(75,79),(80,84),(85,89)]
    dicIntervalos={intervalo:0 for intervalo in intervalos}
    for termo in informacao:
        for intervalo in intervalos:
            if intervalo[0] <= termo[1] <=intervalo[1]:
                dicIntervalos[intervalo] += 1
    max_width = max(len(f"({intervalo[0]}, {intervalo[1]})") for intervalo in intervalos)
    for intervalo,total in dicIntervalos.items():
        p_R=(total/diab)*100
        formatted_range = f"({intervalo[0]}, {intervalo[1]})"
        print(f"                                    {formatted_range:^{max_width}}  | {total} ({p_R:0.2f}%)")

def DistribuicaoGlucose():
    intervalos= [(x, x+9) for x in range(110, 321, 10)]
    dic_intervalos = {intervalo: 0 for intervalo in intervalos}
    for termo in informacao:
        for intervalo in intervalos:
            if intervalo[0] <= termo[7] <= intervalo[1]:
                dic_intervalos[intervalo] += 1
    max_width = max(len(f"({intervalo[0]}, {intervalo[1]})") for intervalo in intervalos)
    for intervalo, total in dic_intervalos.items():
        p_R = (total / diab) * 100
        formatted_range = f"({intervalo[0]}, {intervalo[1]})"

        print(f"                                    {formatted_range:^{max_width}}  | {total} ({p_R:0.2f}%)")

def Table():
    print("""------------------------Análise de dados de uma amostra populacional---------------------------
    ---------------------------Estudo de dados de pacientes diabéticos-----------------------------
    ----------------------------Amostra de 100000 (cem mil) indivíduos-----------------------------
    ----------------------------------8000 (oito mil) pacientes------------------------------------
    ---------Dados em estudo: Distribuição por género, idade e niveis de glucose no sangue---------
    -------------------------------------------Género----------------------------------------------""")
    DistribuicaoGenero()
    print("--------------------------------------------Idade----------------------------------------------")
    DistribuicaoIdade()
    print("----------------------------------Niveis de glucose no sangue-----------------------------------")
    DistribuicaoGlucose()

Table()
```
